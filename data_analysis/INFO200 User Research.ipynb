```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
```

```python
df = pd.read_csv("E:/Jupyter/info200_final_project/survey_raw.csv")
```

```python
def filter_english(s):
    splited = s.split(" (")
    return splited[0]
```

```python
df['Job'] = df['Job'].apply(filter_english)
```

```python
customers = df[df['Job'] != 'Business owner']
business_owners = df[df['Job'] == 'Business owner']
```

```python
customers['Do you use the government app Thai-Chana?'] = customers['Do you use the government app Thai-Chana?'].apply(filter_english)
business_owners['Do you use the government app Thai-Chana?.1'] = business_owners['Do you use the government app Thai-Chana?.1'].apply(filter_english)
```

```python
customers = pd.read_csv("E:/Jupyter/info200_final_project/customers.csv")
customers
```

```python
business_owners = pd.read_csv("E:/Jupyter/info200_final_project/business_owners.csv")
business_owners
```

```python
import matplotlib as mpl
mpl.rcParams['font.size'] = 14
```

```python
ages_and_job = pd.concat([customers.loc[:, ['Age', 'Job']], business_owners.loc[:, ['Age', 'Job']]])
```

```python
# Ages of all respondents
f, ax = plt.subplots(figsize = (6, 6))
groupby_age = ages_and_job.groupby('Age')['Age'].count()
ax.pie(groupby_age, labels=groupby_age.index, autopct='%1.1f%%')
ax.set_title('Ages')
plt.savefig("customers_ages.png")
```

```python
# Jobs of all respondents
f, ax = plt.subplots(figsize = (8, 6))
groupby_job = ages_and_job.groupby('Job')['Job'].count()
plt.pie(groupby_job, labels=groupby_job.index, autopct='%1.1f%%')
ax.set_title('Jobs')
plt.savefig("customers_jobs.png")
```

```python
# Customers' responses by question
mpl.rcParams['font.size'] = 12
f, axs = plt.subplots(nrows=3, ncols=2, figsize = (18, 14))
sns.distplot(a=customers['How comfortable do you feel going out during this Covid-19 situation?'], fit=norm, kde=True, ax=axs[0, 0])
sns.distplot(a=customers['Are you comfortable being in places high population density right now?'], fit=norm, kde=True, ax=axs[0, 1])
sns.distplot(a=customers["How confident do you feel about the shops' and restaurants' hygiene standard in general?"], fit=norm, kde=True, ax=axs[1, 0])
sns.distplot(a=customers['How well do you think shops and restaurants are handling social distancing?'], fit=norm, kde=True, ax=axs[1, 1])
sns.distplot(a=customers['How well are you informed about rules and regulations regarding Covid-19 before getting services from different places?'], fit=norm, kde=True, ax=axs[2, 0])
axs[2,0].set_xlabel("How well are you informed about rules and regulations regarding Covid-19\n before getting services from different places?")
sns.distplot(a=customers['Rate how strong Thai-Chana has made you more confident going out and getting services.'], fit=norm, kde=True, ax=axs[2, 1])
for x in axs:
    for y in x:
        y.set_xticks(np.arange(0, 6, step=1))
plt.savefig("customers_plots.png")
```

```python
# Percentages of non-business owners who use Thai-Chana app
mpl.rcParams['font.size'] = 14
f, ax = plt.subplots(figsize = (6, 6))
groupby_thai_chana = groupby_job = customers.groupby('Do you use the government app Thai-Chana?')['Do you use the government app Thai-Chana?'].count()
plt.pie(groupby_thai_chana, labels=groupby_thai_chana.index, autopct='%1.1f%%')
ax.set_title('Do you use the government app Thai-Chana?')
plt.savefig("customers_thai_chana.png")
```

```python
# Business owners' responses by question
mpl.rcParams['font.size'] = 12
f, axs = plt.subplots(nrows=3, ncols=1, figsize = (9, 17))
sns.distplot(a=business_owners['How severe has this Covid-19 situation affected your business?'], hist_kws={'color':'g'}, fit=norm, kde=False, bins=5, ax=axs[0])
sns.distplot(a=business_owners['How much do you think social distancing is necessary to your business?'], hist_kws={'color':'g'}, kde_kws={'color':'g'}, fit=norm, kde=True, bins=5, ax=axs[1])
sns.distplot(a=business_owners['Rate how helpful Thai-Chana is to you business.'], hist_kws={'color':'g'}, kde_kws={'color':'g'}, fit=norm, kde=True, bins=5, ax=axs[2])
for x in axs:
    x.set_xticks(np.arange(0, 6, step=1))
plt.savefig("business_owners_plots.png")
```

```python
# Percentages of business owners who use Thai-Chana app
f, ax = plt.subplots(figsize = (6, 6))
groupby_thai_chana = groupby_job = business_owners.groupby('Do you use the government app Thai-Chana?')['Do you use the government app Thai-Chana?'].count()
plt.pie(groupby_thai_chana, labels=groupby_thai_chana.index, autopct='%1.1f%%')
ax.set_title('Do you use the government app Thai-Chana?')
plt.savefig("business_owners_thai_chana.png")
```
